```python
os.getcwd()
file_path = './data/CRU_data.csv'
dt = pd.read_csv(file_path, sep=";")
csv_data = pd.DataFrame(dt)

# Preparing columns fro precipitation and months
prep_columns = ['PRECIP_JAN', 'PRECIP_FEB', 'PRECIP_MAR', 'PRECIP_APR', 'PRECIP_MAY', 'PRECIP_JUN',
                'PRECIP_JUL', 'PRECIP_AUG', 'PRECIP_SEP', 'PRECIP_OCT', 'PRECIP_NOV', 'PRECIP_DEC']

# Filtering data for the year 2023 and countries
countries_data = csv_data[(csv_data['YEAR'] == 2023)]
countries = [country for country in countries_data.COUNTRY]

cumsum_precip_list = []

for country in countries:
    # Filtering data for each country and replacing outliers
    data = csv_data[(csv_data['COUNTRY'] == country) & (csv_data['YEAR'] == 2023)].replace(-999, np.nan)

    # Calculating the cumulative sum for the year 2023 for each country
    cumsum_precip = data[prep_columns].dropna().sum(axis=1).values.flatten()
    
    # Storing the cummsum against the country in the lsit
    cumsum_precip_list.append({'Country': country, 'Cumsum_Precip': cumsum_precip[0]})

    # Coverting the list into DataFrame
    precip_df = pd.DataFrame(cumsum_precip_list)

# Plotting data
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(12, 8))
plt.bar(precip_df['Country'], precip_df['Cumsum_Precip'], color='c', alpha=0.5)
plt.xticks(rotation=45)
plt.title('Cumulative Precipitation in 2023')
plt.xlabel('Country')
plt.ylabel('Cumulative Precipitation (mm)')
plt.grid(axis='y', linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

#### The multimodal graph above shows the cumulative precipitation of all countries as well as world wide. Majority of the countries, about 7, lie within the 500mm and 750mm range. Most of these countries are mediterranean countries, Greece, Italy, Lebanon, Spain, and Turkey, which shows that they received about the similar amounts of rainfall in 2023. On the contrary, Egypt received the least rainfall of around 30mm with Iran and Tunisia follwoing. Vietnam and India stand alone in their categories being the two countries with the most cumulative precipitation of 1720mm and 1170mm respectively. France and Germany fall close the to the world-wide precipitation rate of around 830mm.

### Task 3.2

```python
# Defining a list that will contain average percipitation rate per wet day for each country
avg_precip_list = []

# Defining columns for monthly wet days, precipitation columns will be taken from previous cell
wet_day_columns = ['WET_DAYS_JAN', 'WET_DAYS_FEB', 'WET_DAYS_MAR', 'WET_DAYS_APR', 'WET_DAYS_MAY', 'WET_DAYS_JUN', 
                   'WET_DAYS_JUL', 'WET_DAYS_AUG', 'WET_DAYS_SEP', 'WET_DAYS_OCT', 'WET_DAYS_NOV', 'WET_DAYS_DEC']

for country in countries:
    # Filtering data for the specific country and year, and replacing outliers with NaN
    country_data_2023 = csv_data[(csv_data['COUNTRY'] == country) & (csv_data['YEAR'] == 2023)].replace(-999, np.nan)
    
    # Calculating total precipitation and wet days for the country
    total_precipitation = country_data_2023[prep_columns].values.flatten().sum()
    total_wet_days = country_data_2023[wet_day_columns].values.flatten().sum()
    
    # Calculating average precipitation per wet day, only if wet days are more than zero (can't divide by zero)
    avg_precip = total_precipitation / total_wet_days if total_wet_days > 0 else None

    # Adding final precipitation rate for each country in the list and dropping the NaN values
    avg_precip_list.append({'COUNTRY': country, 'Avg_Precip_per_Wet_Day': avg_precip})
    result = pd.DataFrame(avg_precip_list).dropna()
    
# Plotting the result
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(12, 8))
plt.bar(result['COUNTRY'], result['Avg_Precip_per_Wet_Day'], color='pink')
plt.xticks(rotation=60)
plt.title(f'Average Precipitation per Wet Day in 2023')
plt.xlabel('Country')
plt.ylabel('Average Precipitation per Wet Day (mm)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### This graph shows the average precipitation rate per wet day of all the countries included in the dataset. Majority of the countries, around 7, are within the 5.0 mm to 7.5mm range which is also reflective of their geographical location as they are mediterranean countries (except for Australia and China). Two other mediterranean countries (Spain and Egypt) follow suit, falling right below 5.0mm along with Germany and Iran. All South Asian and Southeast Asian countries receive the most rain staying above 7.5mm range with India standing out at 20mm and Vietnam at 10.0mm.

```python
# Preparing all TMP columns
tmp_mean_columns = ['TMP_MEAN_JAN', 'TMP_MEAN_FEB', 'TMP_MEAN_MAR', 'TMP_MEAN_APR', 'TMP_MEAN_MAY', 'TMP_MEAN_JUN', 'TMP_MEAN_JUL', 'TMP_MEAN_AUG', 
         'TMP_MEAN_SEP', 'TMP_MEAN_OCT', 'TMP_MEAN_NOV', 'TMP_MEAN_DEC']

tmp_min_columns = ['TMP_MIN_JAN', 'TMP_MIN_FEB', 'TMP_MIN_MAR', 'TMP_MIN_APR', 'TMP_MIN_MAY', 'TMP_MIN_JUN', 'TMP_MIN_JUL', 'TMP_MIN_AUG', 
         'TMP_MIN_SEP', 'TMP_MIN_OCT', 'TMP_MIN_NOV', 'TMP_MIN_DEC']

tmp_max_columns = ['TMP_MAX_JAN', 'TMP_MAX_FEB', 'TMP_MAX_MAR', 'TMP_MAX_APR', 'TMP_MAX_MAY', 'TMP_MAX_JUN', 'TMP_MAX_JUL', 'TMP_MAX_AUG', 
         'TMP_MAX_SEP', 'TMP_MAX_OCT', 'TMP_MAX_NOV', 'TMP_MAX_DEC']

annual_mean_list = []

for year in range(2014, 2024):
    # Filtering data for the country and each year, and replacing outliers
    data = csv_data[(csv_data['COUNTRY'] == 'PAKISTAN') & (csv_data['YEAR'] == year)].replace(-999, np.nan)

    # Storing specific columns
    mean_tmp, precip = data[tmp_mean_columns], data[prep_columns]
    min_tmp, max_tmp = data[tmp_min_columns], data[tmp_max_columns]

    # Calculating mean values for temperature and precipitation
    mean_tmp_mean = mean_tmp.values.mean()
    min_tmp_mean = min_tmp.values.mean()
    max_tmp_mean = max_tmp.values.mean()
    
    precip_mean = precip.values.mean()

    # Appending all data to the list
    annual_mean_list.append({'Year': year, 'TMP_Mean': mean_tmp_mean, 'TMP_Min': min_tmp_mean, 'TMP_Max':max_tmp_mean, 'PRECIP_Mean': precip_mean})

    # Converting the list to DataFrame and dropping NaN values
    annual_mean = pd.DataFrame(annual_mean_list).dropna()

# Calculating max values for graphs
precip_max = annual_mean['PRECIP_Mean'].max()
tmp_max = annual_mean['TMP_Mean'].max()

# Plotting the bar graph showing precipitation
fig, tmp_ax = plt.subplots(figsize=(9, 7), sharex=False)
tmp_ax.set_xlabel('Year')
tmp_ax.set_ylabel('Mean Precipitation (mm)', color = 'black')
tmp_ax.set_ylim(ymin=0, ymax=precip_max+20)
tmp_ax.set_xticks(annual_mean['Year'])
tmp_ax.tick_params('x', labelrotation=45)
tmp_ax.bar(annual_mean['Year'], annual_mean['PRECIP_Mean'], color='orange', label='Precipitation', alpha=0.7)
tmp_ax.tick_params(axis='y', labelcolor='black')
tmp_ax.grid(axis='y', linestyle='--', alpha=1)

# Plotting the line graph on the same graph showing min, max, and mean temperature
precip_ax = tmp_ax.twinx()
precip_ax.set_ylabel('Temperature (°C)', color = 'black')
precip_ax.set_ylim(ymin=0, ymax=tmp_max+10)
precip_ax.plot(annual_mean['Year'], annual_mean['TMP_Mean'], marker='o', linestyle='-', color='green', label='Mean Temp')
precip_ax.plot(annual_mean['Year'], annual_mean['TMP_Min'], marker='o', linestyle='-', color='blue', label='Min Temp')
precip_ax.plot(annual_mean['Year'], annual_mean['TMP_Max'], marker='o', linestyle='-', color='red', label='Max Temp')
precip_ax.tick_params(axis='y', labelcolor='black')
precip_ax.grid(linestyle='--', alpha=1)

# Adding labels to the graph
plt.rcParams.update({'font.size': 14})
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig.tight_layout()
plt.title('Climate Diagram of ' + data['COUNTRY'].values[0] + ' from 2014 to 2023')
plt.show()
```

```python
# A bit of duplicated code due to division of labour, and to be on a safe side.
import pandas as pd

data = pd.read_csv("./data/CRU_data.csv", sep=";")
```

```python
""" 1. For each month, calculate if it was warmer or colder compared to the same month 50 years ago starting from 
the year 1964 onwards.
"""

# Define monthly columns
monthly_columns = [f'TMP_MEAN_{month}' for month in 
                   ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]]

# Filter rows for Pakistan and Spain from the year 1964 onward, ignoring rows with -999 values
data_filtered = (
    data.query("COUNTRY in ['PAKISTAN', 'SPAIN']")
    .set_index(["COUNTRY", "YEAR"])
    .loc[:, monthly_columns]
    .replace(-999.0, pd.NA)
    .dropna()
)

# Initialize an empty DataFrame to store comparison results
comparison = pd.DataFrame(columns=["Country", "Year", "Month", "Change"])

# Loop through each year starting from 1964
for year in range(1964, data_filtered.index.get_level_values("YEAR").max() + 1):
    comparison_year = year - 50
    # if comparison_year < data_filtered.index.get_level_values("YEAR").min():
    #     continue  # Skip years without a 50-year prior comparison

    # Extract data for the current year and the comparison year
    data_current = data_filtered.xs(year, level="YEAR")
    data_comparison = data_filtered.xs(comparison_year, level="YEAR")

    # Compare each month and determine if it's warmer, colder or no change
    change = data_current.subtract(data_comparison)
    change_mask = change.apply(
        lambda col: col.map(lambda x: "Warmer" if x > 0 else "Colder" if x < 0 else "No Change"))

    # Melt the change data into a long format
    monthly_comparison = change_mask.reset_index()
    monthly_comparison["Year"] = year
    monthly_comparison = monthly_comparison.melt(id_vars=["COUNTRY", "Year"], var_name="Month", value_name="Change")
    monthly_comparison["Month"] = monthly_comparison["Month"].str.replace("TMP_MEAN_", "")
    monthly_comparison = monthly_comparison.rename(columns={"COUNTRY": "Country"})

    # Append the results to the comparison DataFrame
    comparison = pd.concat([comparison, monthly_comparison], ignore_index=True)

# Capitalize country names and add 'Decade' column for further grouping
comparison["Country"] = comparison["Country"].str.capitalize()
comparison["Decade"] = (comparison["Year"] // 10) * 10

comparison
```

```python
""" 2.1. Absolute counts of warmer and colder months per country with row and column margins. """

contingency_absolute_country = pd.crosstab(
        index=comparison["Country"], columns=comparison["Change"], margins=True
    ).rename(columns={"All": ""}, index={"All": ""})


# Ensure there's no additional header by setting both index and columns names to None
contingency_absolute_country.columns.name = None
contingency_absolute_country.index.name = None

contingency_absolute_country
```

```python
""" 2.2. Proportions of warmer and colder months per country with row and column margins. """

# Proportion contingency table (row and column proportions)
contingency_proportion_country = pd.crosstab(
        comparison["Country"], comparison["Change"], normalize="all", margins=True
    ).rename(columns={"All": ""}, index={"All": ""})

# Ensure there's no additional header by setting both index and columns names to None
contingency_proportion_country.columns.name = None
contingency_proportion_country.index.name = None

contingency_proportion_country.round(3)
```

```python
""" 3.1. Absolute counts of warmer and colder months per decade with row and column margins. """

# Contingency table with absolute counts per decade
contingency_absolute_decade = pd.crosstab(
        comparison["Decade"], comparison["Change"], margins=True
    ).rename(columns={"All": ""}, index={"All": ""})

# Ensure there's no additional header by setting both index and columns names to None
contingency_absolute_decade.columns.name = None
contingency_absolute_decade.index.name = None

contingency_absolute_decade
```

```python
""" 3.2. Proportions of warmer and colder months per country with row and column margins. """

# Proportion contingency table (row and column proportions)
contingency_proportion_decade = pd.crosstab(
        comparison["Decade"], comparison["Change"], normalize="all", margins=True
    ).rename(columns={"All": ""}, index={"All": ""})

# Ensure there's no additional header by setting both index and columns names to None
contingency_proportion_decade.columns.name = None
contingency_proportion_decade.index.name = None

contingency_proportion_decade.round(3)
```

```python
""" 4. A bar chart of total number of warmer months by country and decade. """

import matplotlib.pyplot as plt

# Set the font size globally
plt.rcParams.update({'font.size': 14})

# Filter only "Warmer" status for plotting
warmer_data = comparison[comparison["Change"] == "Warmer"]

# Group by Decade and Country, count the occurrences of "Warmer" months
warmer_counts = warmer_data.groupby(["Decade", "Country"]).size().unstack(fill_value=0)

# Plot the data
fig, ax = plt.subplots(figsize=(12, 7))
warmer_counts.plot(kind="bar", ax=ax, stacked=False, color=["#1f77b4", "#ff7f0e"])

# Title and labels
ax.set_title("Total Number of Warmer Months by Country and Decade")
ax.set_xlabel("Decade")
ax.set_ylabel("Number of Warmer Months")
plt.xticks(rotation=45)
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

#### Interpretation

The bar chart illustrates the total number of warmer months compared to a baseline 50 years ago for Pakistan and Spain, tracked across each decade from the 1960s to the 2020s. The chart reveals an **overall warming trend** in both countries with the number of warmer months increasing from the 1960s through the 2010s, with the 1990s appears to be an exception to the general pattern, possibly due to natural climate variability or large-scale climate phenomena.
```python
""" 5. Combined contingency tables with a subdivision for countries by decade,
with absolute counts and row/column proportions.
"""

# Absolute counts by country and decade with countries as the primary level
contingency_combined_absolute = pd.crosstab(
    [comparison["Country"], comparison["Decade"]], comparison["Change"], margins=True
)

# Proportion contingency table by country and decade
contingency_combined_proportion = pd.crosstab(
    [comparison["Country"], comparison["Decade"]], comparison["Change"], normalize="all", margins=True
)

# Combining the absolute and proportion contingency tables into a single DataFrame for easier comparison
# Renaming columns for clarity in combined output
contingency_combined_absolute = contingency_combined_absolute.add_suffix(" Count")
contingency_combined_proportion = contingency_combined_proportion.add_suffix(" Proportion")

# Joining the two tables
contingency_combined = contingency_combined_absolute.join(contingency_combined_proportion)

# Remove the index and column names for a cleaner look
contingency_combined.index.names = [None, None]
contingency_combined.columns.name = None  # Remove the "Change" header

# Reorder columns to move "All Count" to the end
columns = [col for col in contingency_combined.columns if col != "All Count"] + ["All Count"]
contingency_combined = contingency_combined[columns]

def highlight_all_count(col):
    color = "background-color: #E0E0E0"
    return [color if col.name in ("All Count", "All Proportion") else "" for _ in col]

# Identify only the proportion columns for formatting
proportion_columns = [col for col in contingency_combined.columns if "Proportion" in col]

# Apply styling to highlight the "All Count" column in grey
contingency_combined = contingency_combined.style.apply(
    highlight_all_count, subset=["All Proportion", "All Count"]
).format("{:.3f}", subset=proportion_columns)

contingency_combined
```

```python
""" 6. The expected frequencies for each conjunct event in the joint contingency table. """

# Create the absolute counts and proportion contingency tables with margins
contingency_combined_absolute = pd.crosstab(
    [comparison["Country"], comparison["Decade"]], 
    comparison["Change"], 
    margins=True
)

contingency_combined_proportion = pd.crosstab(
    [comparison["Country"], comparison["Decade"]], 
    comparison["Change"], 
    normalize="all", 
    margins=True
)

# Add suffixes to distinguish absolute counts, proportions, and expected frequencies
contingency_combined_absolute = contingency_combined_absolute.add_suffix(" Count")
contingency_combined_proportion = contingency_combined_proportion.add_suffix(" Proportion")

# Calculate expected frequencies using row and column totals from the absolute counts table
row_totals = contingency_combined_absolute.iloc[:-1, -1]       # Row totals, excluding the last row
column_totals = contingency_combined_absolute.iloc[-1, :-1]    # Column totals, excluding the last column
grand_total = contingency_combined_absolute.iloc[-1, -1]       # Overall grand total

# Create DataFrame to store expected frequencies without margins
expected_frequencies = pd.DataFrame(index=row_totals.index, columns=column_totals.index)

# Populate expected frequencies table
for row in expected_frequencies.index:
    for col in expected_frequencies.columns:
        expected_frequencies.loc[row, col] = (row_totals[row] * column_totals[col]) / grand_total

# Add suffix to expected frequencies for clarity
expected_frequencies = expected_frequencies.add_suffix(" Expected")

# Combine all tables into one DataFrame
contingency_combined = pd.concat(
    [contingency_combined_absolute, contingency_combined_proportion, expected_frequencies], axis=1)

# Round proportions and expected frequencies for readability
proportion_columns = [col for col in contingency_combined.columns if "Proportion" in col]
expected_columns = [col for col in contingency_combined.columns if "Expected" in col]
contingency_combined[proportion_columns] = contingency_combined[proportion_columns].astype(float).round(3)
contingency_combined[expected_columns] = contingency_combined[expected_columns].astype(float).round(3)

# Clean up index and column names for readability
contingency_combined.index.names = [None, None]
contingency_combined.columns.name = None

# Order columns to keep "All" totals at the end
ordered_columns = (
    [col for col in contingency_combined.columns if "All" not in col] + 
        [col for col in contingency_combined.columns if "All" in col])
contingency_combined = contingency_combined[ordered_columns]

contingency_combined
```

```python
""" 7. Chi-squared values for Pakistan and Spain. """

# Initialize chi-squared values for Pakistan and Spain
chi2_pakistan = 0
chi2_spain = 0

# Define the columns for observed and expected values based on the data structure
observed_columns = ["Colder Count", "Warmer Count", "No Change Count"]
expected_columns = ["Colder Count Expected", "Warmer Count Expected", "No Change Count Expected"]

# Loop over each observed and expected column pair
for observed_col, expected_col in zip(observed_columns, expected_columns):
    # Iterate over each decade in the index
    for decade in contingency_combined_absolute.index.get_level_values("Decade").unique():
        # Skip the "All" row in the index
        if decade == "":
            continue

        # Calculate chi-squared for Pakistan
        observed_pakistan = contingency_combined_absolute.loc[("Pakistan", decade), observed_col]
        expected_pakistan = expected_frequencies.loc[("Pakistan", decade), expected_col]
        if expected_pakistan > 0:
            chi2_pakistan += ((observed_pakistan - expected_pakistan) ** 2) / expected_pakistan

        # Calculate chi-squared for Spain
        observed_spain = contingency_combined_absolute.loc[("Spain", decade), observed_col]
        expected_spain = expected_frequencies.loc[("Spain", decade), expected_col]
        if expected_spain > 0:
            chi2_spain += ((observed_spain - expected_spain) ** 2) / expected_spain

print("Chi-squared value for Pakistan:", round(chi2_pakistan, 3))
print("Chi-squared value for Spain:", round(chi2_spain, 3))
```

#### Interpretation
The calculated $\chi^2$ values for Pakistan and Spain indicate how much the observed changes in temperature deviate from expected. A high $\chi^2$ value suggests that there is a significant difference between the observed and expected frequencies. A low $\chi^2$ value would suggest that the observed values are close to what we would expect by chance, indicating no significant deviation.

Pakistan has a higher $\chi^2$ value (73.574) compared to Spain (41.335), indicating that the observed data for Pakistan deviates more from the expected counts than it does for Spain. This suggests that the climate or temperature changes in Pakistan are further from the expected baseline (based on historical data or another baseline comparison) than they are in Spain. This might indicate more pronounced temperature trends or climate variability in Pakistan relative to Spain.

#### 8. What does a small chi-square value mean? What if it's zero?
A small $\chi^2$ value indicates that there is little difference between the observed and expected frequencies in a chi-squared test.
In hypothesis testing, a small $\chi^2$ value, especially one approaching zero, can lead to a conclusion that the factors under consideration do not have a statistically significant impact on the outcome being measured.

If the $\chi^2$ value is small enough to be below the critical value from the chi-squared distribution for the given degrees of freedom and significance level (usually 0.05), it means that you fail to reject the null hypothesis. While a $\chi^2$ value of zero is theoretically possible, it may also indicate that your model or test is overly simplistic or not accounting for variability in the data. It could be a sign that there is not enough data or diversity in the sample to show any deviation from expectations.
